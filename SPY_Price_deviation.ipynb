```python
import matplotlib.pyplot as plt
import pandas as pd
import yfinance as yf

# get data
spy = yf.Ticker("SPY")
SpyData = spy.history(period="5y") 
#print(SpyData)

# Create the histogram
plt.hist(SpyData['Close'], bins=30, edgecolor='black') 

# Add labels and title
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.title('SPY Prices over 5 years')

# Show the plot
plt.show()
```

```python
SpyData2 = spy.history(period="5y") 

# Find change in price with .diff() from Pandas
SpyData2['Price Change'] = SpyData['Close'].diff()

# Removes Null values with .dropna
SpyData2 = SpyData2.dropna()

plt.hist(SpyData2['Price Change'], bins=50, edgecolor='black')

plt.xlabel('Daily Price Change')
plt.ylabel('Frequency')
plt.title('Distribution of Daily SPY Price Changes Over 5 Years')

plt.show()

```

```python
SpyData3 = spy.history(period='5y')

SpyData3['Percent Change'] = SpyData3['Close'].pct_change() * 100

SpyData3 = SpyData3.dropna()

plt.hist(SpyData3['Percent Change'], bins=20, edgecolor='black')


plt.xlabel('Daily Percent Change')
plt.ylabel('Frequency')
plt.title('Distribution of SPY Percent Changes Over 5 Year')

plt.show()
```

```python
SpyData4 = spy.history(period='5y')

#weekly SPY deviation 
SpyData4 = SpyData4.resample('W').last()

SpyData4['Percent Change'] = SpyData4['Close'].pct_change() * 100

SpyData4 = SpyData4.dropna()

plt.figure(figsize=(10, 6))

plt.hist(SpyData4['Percent Change'], bins=20, edgecolor='black')


plt.xlabel('Percent Change')
plt.ylabel('Frequency')
plt.title('Distribution of SPY Percent Changes Over 5 Year (Weekly)')

plt.show()
```
